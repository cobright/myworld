```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import mglearn

db = load_iris()
db.feature_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(db['data'],db['target'], random_state=0)
```

```python

df = pd.DataFrame(X_train, columns=db.feature_names)
pd.plotting.scatter_matrix(df, c=y_train, figsize=(15,15), marker='o', hist_kwds={'bins':20}, s=60, alpha=.8,cmap=mglearn.cm3)
```
